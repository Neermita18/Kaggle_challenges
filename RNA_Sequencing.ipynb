## Assessing Sequence Similarities

```python
def sim(seq1, seq2):
  alignments = pairwise2.align.globalxx(seq1, seq2)
  # print(alignments)
  # print(type(alignments[0]))
  s= alignments[0]
  # print(s[0])
  # print(s[1])
  s1=s[0]
  s2=s[1]
  c=0
  d=0
  l= max(len(s[0]), len(s[1]))
  for i in range(0,l):
    if s1[i]==s2[i]:
      c+=1
      d+=1
    else:
      d+=1
  return c/d
s1= Seq(df.iloc[0,1])
s2= Seq(df.iloc[10,1])
sim(s1,s2 )
```

```python
for i in range(1,40):
  s1= Seq(df.iloc[0,1])
  s2= Seq(df.iloc[i,1])
  x=sim(s1,s2)
  print(f"Similarity between seq 1 and seq {i}: {x}")
```
